### Tabla códigos municipios

```python
# Tengo 52 hojas en el fichero de Excel que contiene los códigos de los municipios, una por cada municipio
cod_muni = pd.DataFrame()
for muni in range(52):
    cod_muni= pd.concat([cod_muni,pd.read_excel('./datos/MIGRACIONES/23codmun.xlsx', sheet_name=muni, header=2)], ignore_index=True)
cod_muni
```

```python
cod_muni.columns
```

```python
# Elimino posibles espacios antes o después del nombre de las columnas
cod_muni.columns = list(map(lambda x: x.strip(),list(cod_muni.columns)))
```

```python
# Los códigos de municipio completos con los que se trabajan habitualmente son
# la suma del código de provicia y el código de municipio de la tabla
```

```python
# Creo una nueva columna con este código

cod_muni['CMUNICIPIO'] = cod_muni.apply(lambda x: f"{x['CPRO']:02}" + f"{x['CMUN']:03}", axis=1)
                                 
cod_muni
```

```python
cod_muni[cod_muni['CPRO']== 46]
```

```python
# Añado una fila con el valor CPRO 66, CMUN 666 y CMUNICIPIO 66666 que corresponde a 'extranjero'
# y fila con valores 0 como descondocido

cod_muni.loc[len(cod_muni.index)] = [66, 6, 0, 'Extranjero', '66666']
cod_muni.loc[len(cod_muni.index)] = [0, 0, 0, 'Desconocido', '00000']
cod_muni
```
